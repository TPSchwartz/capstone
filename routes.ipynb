```python
tn_routes = pd.read_json('C:/Users/cheft/Desktop/NSS/Capstone/capstone/data/tn-routes.jsonlines', lines=True)
tn_routes
```

```python
tn_areas = pd.read_json('C:/Users/cheft/Desktop/NSS/Capstone/capstone/data/tn-areas.jsonlines', lines=True)
tn_areas
```

```python
# Replace grade dict with only YDS grades

yds = []
for index, row in tn_routes.iterrows():
    yds.append(tn_routes.grade[index].get('YDS'))

tn_routes['grade'] = yds
```

```python
# Turn type dictionaries into type lists
string = ''
types = []
for index, row in tn_routes.iterrows():
    types.append(list(tn_routes.type[index].keys()))
tn_routes['type'] = types
```

```python
tn_routes.metadata[0]
```

```python
# Pull parent_sector out of metadata into its own column

parents = []
for index, row in tn_routes.iterrows():
    parents.append(tn_routes.metadata[index].get('parent_sector'))

tn_routes['area'] = parents

# Repeat for route_id

ids = []
for index, row in tn_routes.iterrows():
    ids.append(tn_routes.metadata[index].get('mp_route_id'))

tn_routes['route_id'] = ids

# Repeat for sector_id

sectors = []
for index, row in tn_routes.iterrows():
    sectors.append(tn_routes.metadata[index].get('mp_sector_id'))

tn_routes['area_id'] = sectors

# Repeat for longlat

coords = []
for index, row in tn_routes.iterrows():
    coords.append(tn_routes.metadata[index].get('parent_lnglat'))

tn_routes['area_lnglat'] = coords
```

```python
# Drop safety and metadata column

tn_routes = tn_routes.drop(columns=['metadata','safety'])
```

```python
# Make protection lists into strings

protection = []
for index,row in tn_routes.iterrows():
    protection.append(''.join(str(i) for i in tn_routes.protection[index]))
tn_routes['protection'] = protection

# Repeat for type

typ = []
for index,row in tn_routes.iterrows():
    typ.append(''.join(str(i) for i in tn_routes.type[index]))
tn_routes['type'] = typ

# Repeat for description

descr = []
for index,row in tn_routes.iterrows():
    descr.append(''.join(str(i) for i in tn_routes.description[index]))
tn_routes['description'] = descr

# Repeat for location

loca = []
for index,row in tn_routes.iterrows():
    loca.append(''.join(str(i) for i in tn_routes.location[index]))
tn_routes['location'] = loca
```

```python
# Pull area ID out of url
aid = []
for index, row in tn_areas.iterrows():
    aid.append(re.findall(r'\d+',tn_areas.url[index])[0])

tn_areas['area_id'] = aid
```

```python
# Pull lnglat into separate columns

lng = []
for index, row in tn_areas.iterrows():
    lng.append(tn_areas.lnglat[index][0])

lat = []
for index, row in tn_areas.iterrows():
    lat.append(tn_areas.lnglat[index][1])

# Add longitude and latitude columns

tn_areas['longitude'] = lng
tn_areas['latitude'] = lat
```
