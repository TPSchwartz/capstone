```python
# Replace grade dict with only YDS grades

yds = []
for index, row in tn_routes.iterrows():
    yds.append(tn_routes.grade[index].get('YDS'))

tn_routes['grade'] = yds
```

```python
# Turn type dictionaries into type lists

types = []
for index, row in tn_routes.iterrows():
    types.append(list(tn_routes.type[index].keys()))
tn_routes['type'] = types
```
